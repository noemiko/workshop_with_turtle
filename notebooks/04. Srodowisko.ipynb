Jednak dla celów warsztatowych o wiele łatwiej jest korzystać z notatników w Jupyterze.

Jupyter to środowisko podobne do notatnika. Pozwala on jednak wykonywać kod oraz wizualizować wyniki. Jest podobne do skoroszytu w excelu, bo też posiada komórki, które mogą wykonywać obliczenia.

![excel.png](images/excel.png)

W excelu możemy wyliczyć zawartośc komórki przez kliknięcie Enter. W jupyterze jest parę możliwości uruchomienia komórki




Wyniki obliczeń wyświetlają się pod komórką. Liczba obok komórki informuje kolejność wykonania komórki.

![sekwencje.png](images/sekwencje.png)

Gwiazdka zamiast numeru kolejności wykonania komórki świadczy o tym, że kod jeszcze jest wykonywany

![gwiazdka](images/notebook_gwiazdka.png)

Po warsztacie możecie pracować na waszych plikach na https://colab.research.google.com/
    
![download.png](images/how_download.png)
